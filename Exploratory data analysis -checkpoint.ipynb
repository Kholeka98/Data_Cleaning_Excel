# Step 2: Data Preparation
- Dropping irrelevant columns and rows
- Identifying duplicated columns
- Renaming Columns


```python
#Dropping irrelevant columns
df1 = df[[
        'Product type', 'SKU', 'Price', 'Availability','Number of products sold', 'Revenue generated', 'Customer demographics',
        #'Stock levels', 'Lead times', 
        'Order quantities', 
        #'Shipping times', 'Shipping carriers', 
        'Shipping costs', #'Supplier name', 
        'Location',
        #'Lead time', 'Production volumes', 'Manufacturing lead time',
        'Manufacturing costs', 
        #'Inspection results', 'Defect rates', #'Transportation modes', 'Routes', 'Costs'
        ]].copy()

# Display first 5 rows of the new dataset
df1.head()
```

```python
#Returns duplicate rows
df1.loc[df1.duplicated()]
```

This shows that there are no duplicates in our dataset.

```python
#Describe the sliced dataset
df1.describe()
```

```python
#Rename columns to include units
df1 = df1.rename( columns= {'Price':'Price($)', 'Revenue generated': 'Revenue generated($)',
                     'Shipping costs': 'Shipping costs($)', 'Manufacturing costs':'Manufacturing costs($)'})
```

# Step 3: Feature Understanding
(Univariate analysis)

- Plotting Feature Distributions
    - Bargraph
    - KDE
    - Boxplot

```python
#Create a data with product type and price only for plotting
product_price = df1[['Product type','Price($)']]

#Plot the boxplot
sn.boxplot(x= 'Product type', y= 'Price($)', data= product_price)
plt.title("Price Distribution Across Different Product Types ")
plt.show()
```

After analyzing the boxplots for skincare, hair care, and cosmetics, the following observations were made:

**Skincare:**
The boxplot for skincare exhibits a wide box, suggesting a higher interquartile range.
This wider spread indicates a greater variability in prices within the skincare product category.

**Hair Care and Cosmetics:**
In comparison, both hair care and cosmetics show approximately equal spreads, as evidenced by the boxplot widths.
However, cosmetics have a higher median compared to hair care.

**Distribution Skewness:**
The distribution of skincare prices appears to be skewed to the right, implying that there are relatively more high-priced items in the skincare category. There seem to be no skew in the haircare category.
Finally, the cosmetics distribution also appears to slightly skewed to the left, implying that there are more low-priced items in the cosmetics category.

```python
# Creating bar plot
plt.bar(x= 'Product type', height = 'Revenue generated($)',data = df1)
plt.title('Product type revenue breakdown')
plt.show()
```

The revenue generated from haircare, skincare, and cosmetics is roughly equivalent. However, within these categories, haircare stands out with the highest revenue.

# Step 4: Answer the following questions.


**Question 1:** What is the revenue generated from each location?

```python
df1.Location.unique()
```

```python
df1.Location.value_counts()
```

```python
df1.groupby('Location')['Revenue generated($)'].agg('sum')
```

```python
#Calculates total revenue
Total_revenue = df1['Revenue generated($)'].sum()
Total_revenue
```

```python
#Express revenue generated in each location by percentage
df2 = df1.groupby('Location')['Revenue generated($)'].agg('sum').apply( lambda x : x*100/Total_revenue)
df2
```

From the above analysis, we observe that the revenue generated from product sales is distributed as follows:
- 17.76% from Bangalore,
- 20.62% from Chennai,    
- 14.03% from Delhi,         
- 23.73% from Kolkata, and   
- 23.85% from Mumbai.     

**Question 2:** What is the relationship between price and the number of products sold for different product types?

```python
sn.scatterplot(x= 'Price($)', y="Number of products sold",data =df1[df1['Product type']=='haircare'])
plt.title("Relationship Between Haircare Product Sales Volume and Pricing")
plt.show()

sn.scatterplot(x= 'Price($)', y="Number of products sold",data =df1[df1['Product type']=='skincare'])
plt.title("Relationship Between Skincare Product Sales Volume and Pricing")
plt.show()

sn.scatterplot(x= 'Price($)', y="Number of products sold",data =df1[df1['Product type']=='cosmetics'])
plt.title("Relationship Between Cosmetic Product Sales Volume and Pricing")
plt.show()
```

**Analysis of Sales Volume and Pricing in Haircare, Skincare, and Cosmetics Products**

Upon analyzing the relationship between the number of goods sold and the price of haircare and skincare products, it is evident that there is no discernible correlation. This suggests that the number of goods sold is not significantly influenced by the price of these products.

Contrastingly, when examining cosmetics products, a clear negative relationship between sales volume and pricing emerges. Specifically, as the price of cosmetics increases, the corresponding number of goods sold decreases. This inverse correlation implies that higher pricing negatively impacts the demand for cosmetics products, leading to reduced sales volumgory. These insights could inform pricing strategies and marketing efforts aimed at maximizing sales in the respective product categories.

**Question 3**: Which product type generates the most revenue?

```python
df3 = pd.DataFrame(df1.groupby('Product type')['Revenue generated($)'].agg('sum'))
df3['Revenue generated (%)'] = df3['Revenue generated($)'].apply(lambda x : x*100/Total_revenue)
df3
```

According to the table above, it can be inferred that  skincare generate the highest revenue, accounting for 41.83 % of the total revenue. However, may we assert this proposition for every geographical location?

```python
df4 = pd.DataFrame(df1.groupby(['Product type','Location'])['Revenue generated($)'].agg('sum'))
df4['Revenue generated (%)'] = df4['Revenue generated($)'].apply(lambda x : x*100/Total_revenue)
df4
```

```python
sn.scatterplot(x= 'Location', y= 'Revenue generated (%)',hue= 'Product type',data= df4)
plt.title("Comparing Percentage Revenue and Location Across Various  Products")
plt.show()
```

Upon analyzing the data, it is evident that the distribution of revenue across different product categories varies among the observed cities. Specifically:

- In Bangalore, the predominant revenue comes from haircare products.
- Chennai's revenue is primarily driven by skincare products.
- The highest revenue in Delhi is attributed to cosmetics.
- Skincare products contribute significantly to revenue in Kolkata.
- Mumbai's revenue is predominantly generated from cosmetics sales.
